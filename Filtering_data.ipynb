Filtering Data (based on multiple conditions) ❗

```python
import pandas as pd #importing the pandas library 
df = pd.read_csv("/content/sample_data/data_set.csv")
```

```python
#Will only show pokemons having Type-1 as Grass
df.loc[(df["Type 1"] == "Grass")]
```

```python
#Will only show pokemons having Type-1 as Grass and Type-2 as Poison
df.loc[(df["Type 1"] == "Grass") & (df["Type 2"] == "Poison")]
```

```python
#Will only show pokemons having Type-1 as Grass, Type-2 as Poison and HP greater than 100
df.loc[(df["Type 1"] == "Grass") & (df["Type 2"] == "Poison") & (df["HP"] > 100)]
```

We can even make another variable as new_df and make write : `new_df = df.loc[(df["Type 1"] == "Grass") & (df["Type 2"] == "Poison") & (df["HP"] > 100)]` 
Here new_df will be modified whereas df will still remain the data as a whole.

```python
new_df = df.loc[(df["Type 1"] == "Grass") & (df["Type 2"] == "Poison") & (df["HP"] > 100)] 
```

```python
#resetting the index and removing the old indexes of our modified data
new_df = new_df.reset_index(drop = True)
new_df
```

```python
#we can also do this in place inorder to save memory
new_df = new_df.reset_index(drop = True, inplace = True)
new_df
```
