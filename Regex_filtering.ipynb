Regex Filtering (filter based on textual patterns) ðŸ†Ž

```python
import pandas as pd
df = pd.read_csv("/content/sample_data/data_set.csv")
```

```python
#Will filter out all pokemon results having 'Mega' in their names
df.loc[df["Name"].str.contains('Mega')]
```

```python
#Will filter out all pokemon results NOT having 'Mega' in their names
df.loc[~df["Name"].str.contains('Mega')]
```

```python
#Will filter out all pokemon results having 'Type 1' as fire or grass
#Importing regular expressions
import re
df.loc[df["Type 1"].str.contains('Fire | Grass', regex=True)]
#If we edit the above statement as the following and add flags=re.I then it will work without the capitalisation
df.loc[df["Type 1"].str.contains('fire|grass', flags=re.I, regex=True)]
```

```python
#Will show all pokemons whose names start with 'Pi' like Pikachu
df.loc[df["Name"].str.contains('^pi[a-z]*', flags=re.I, regex=True)]
```
