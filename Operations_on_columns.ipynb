Making changes to the data ‚å®

```python
import pandas as pd
df = pd.read_csv("/content/sample_data/data_set.csv")

#Creating a column which is total of all parameter based columns named - Total
df["Total"] = df["HP"] + df["Attack"] + df["Defense"] + df["Sp. Atk"] + df["Sp. Def"] + df["Speed"]
df.head(5) #We will see the addition of a column Total in the end of the table
```

```python
#Dropping columns
df = df.drop(columns=["Legendary"])
```

```python
#Adding columns
df["Total"] = df.iloc[:,4:10].sum(axis=1) #If we want sum to be vertical we specify 1 if we want it horizontal we specify 0
```

```python
#Making the columns as a list value
cols = list(df.columns.values)
df = df[cols[0:4] + [cols[-1]] + cols[4:12]]
df.head(5)
```
