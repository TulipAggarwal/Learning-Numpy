Conditional Changes ⚡



```python
import pandas as pd
df = pd.read_csv("/content/sample_data/data_set.csv")
```

```python
#Changing the fire type as Flamer
df.loc[df['Type 1'] == "Fire", "Type 1"] = 'Flamer'
df
```

```python
#Changing the fire type pokemons as Legendary too
df.loc[df['Type 1'] == "Fire", "Legendary"] = True
df
```

```python
#Changing multiple parameters at the same time
#Here all pokemons having Speed > 50 will make the generation and legendary value as Test Value
df.loc[df["Speed"] > 50, ["Generation", "Legendary"]] = 'TEST VALUE'
df
```

```python
#We can also do this as:
#Here all pokemons having Speed > 50 will make the generation and legendary value as Test1 and Test2 
df.loc[df["Speed"] > 50, ["Generation", "Legendary"]] = 'Test1', 'Test2'
df
```
