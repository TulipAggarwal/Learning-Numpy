```python
#Sorting the data in descending order alphabetically
import pandas as pd
df = pd.read_csv("/content/sample_data/data_set.csv")
df.sort_values("Name", ascending=False)
```

```python
#Sorting in Descending order using two parameters
df.sort_values(["Name","HP"], ascending=False)
```

```python
#Sorting in both ascending and descending order using two parameters
df.sort_values(["Name" , "HP"] , ascending = [1,0]) #Here name will be in ascending order and HP in descending order
```
