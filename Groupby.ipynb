Aggregate Statistics using Groupby (Sum, Mean, Counting) üìÅ



```python
import pandas as pd
df = pd.read_csv("/content/sample_data/data_set.csv")
```

```python
#Means of Type of Pokemons
df.groupby(["Type 1"]).mean()
```

```python
#Means of Type of Pokemons in a way that it gets in descending order of Defense
df.groupby(["Type 1"]).mean().sort_values("Defense", ascending = False)
```

```python
df.groupby(["Type 1"]).mean().sort_values("HP", ascending = False)
```

```python
# Will output the number of count of each type of pokemons
df.groupby(["Type 1"]).count()
```

```python
# Just to get the Count column
df["count"] = 1
df
df.groupby(["Type 1", "Type 2"]).count()["count"]
```
